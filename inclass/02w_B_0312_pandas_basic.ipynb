## 날짜를 위한 함수 date_range()

```python
dates = pd.date_range("20250101", periods=6)
dates
```

```python
dates[3]
```

```python
dates = pd.date_range("20250101", end='20250110')
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='3D')
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='3H')
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='W')
dates # 첫주 월요일부터
```

```python
dates = pd.date_range("20250101", periods=6, freq='W-TUE') #화요일출력
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='ME') # 달의 마지막 날
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='MS') # 달의 첫날
dates
```

```python
dates = pd.date_range("20250302", periods=6)
dates
```
