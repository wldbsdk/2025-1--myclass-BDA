```python
p25, p50, p75 = mpg.hwy.quantile([.25, .5, .75])
```

```python
p25
```

```python
p50
```

```python
p75
```

```python
iqr = p75 - p25
iqr
```

```python
# upper fence
UF = p75 + 1.5 * iqr
UF
```

```python
# lower fence
LF = p25 - 1.5 * iqr
LF
```

```python
mpg.hwy.describe()
```

```python
mpg.hwy.describe()['50%']
```

```python
mpg[mpg.hwy > LF][['hwy']]
```

```python
mpg[mpg.hwy < LF]
```

```python
mpg.info()
```

```python
mpg.isna().sum()
```

```python
mpg[ (mpg['hwy'] < 4.5) | (mpg['hwy'] > 40.5 )]
```

```python
mpg[ (mpg['hwy'] < LF) | (mpg['hwy'] > UF )]
```

```python
mpg['hwy'] = np.where((mpg.hwy < LF) | (mpg.hwy > UF), np.nan, mpg['hwy'])
mpg.hwy.isna().sum()
```

```python
mpg[ mpg.hwy.isna() ]
```

```python
len(mpg)
```

```python
dmpg = mpg.dropna(subset = ['hwy'])
len(dmpg)
```

```python
dmpg.groupby('drv').agg(mean_hwy = ('hwy', 'mean'))
```
